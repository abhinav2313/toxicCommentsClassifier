```python
import pandas as pd
from toxic_funcs import Cleaner
import numpy as np
```

```python
#reading only first n for testing purposes
n = 10000
df = pd.read_csv("data/train.csv").set_index('id')
df = df.head(n)
```

## Preprocess

```python
#running through the custom cleaner

clean_comments = []
for comment in df.comment_text.values:
    clean_comments.append(Cleaner(comment))
```

```python
#slapping it onto df

df['cleaned'] = clean_comments
```

```python
from keras.preprocessing.text import Tokenizer
#split texts into words

tokenizer = Tokenizer(num_words = 10000)
```

```python
#
tokenizer.fit_on_texts(df.cleaned.values)
```

```python
tokenized = tokenizer.texts_to_matrix(df.cleaned.values)
```

```python
df
```

# Build labels and features

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical

X = tokenized
y = np.dstack((df.toxic.values, df.severe_toxic.values, df.obscene.values, df.threat.values, df.insult.values , df.identity_hate.values))[0]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1) #stratify fails with low valued categories
'''X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)'''
```

```python
print(f"x-{X_train.shape} y-{y_train.shape}")
y_train
```

# Build model

```python
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Dropout

# Create a Neural Network model here
# YOUR CODE HERE
model = Sequential()
model.add(Dense(100, activation = "relu", input_dim = X_train.shape[1]))
model.add(Dense(50,activation = "relu"))
model.add(Dropout(.1))
model.add(Dense(10, activation = "relu"))
model.add(Dense(y_train.shape[1], activation = "sigmoid"))
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs = 4, verbose=1)
```

```python
model_loss, model_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f"Normal Neural Network Test - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

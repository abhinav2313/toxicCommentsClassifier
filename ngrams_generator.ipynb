```python
from re import subn
from nltk.corpus import stopwords
from autocorrect import spell


def two_grammer(text):
    '''makes 2-grams out of cleaned, spell-checked text'''
    
    #
    
    clean, _ = subn(r'[.,(){}]:;', ' ', text) # removing symbols left in in cleaning function
    clean, _ = subn(r'\s\s+', ' ', clean) # sets of spaces can cause problems
    split = clean.split(' ')
    #print(split)
    #print('----------------------------------------')
    
    stops = set(stopwords.words("english"))
    
    spelt_gram = []
    last_word = ''
    spelt_gram_string = ''
    for word in split:
        if len(word)>0 and len(word)<20 and word not in stops:
            #print(word, end = '-')
            spelt = spell(word)
            #print(spelt, end = '  ')
            spelt_gram.append(last_word + spelt)
            last_word = spelt

    #print('----------------------------------------')

    #print(spelt_gram)
    #print('----------------------------------------')
    #print(text)
    
    spelt_gram_string = ''
    for pair in spelt_gram[1:]:
        spelt_gram_string += pair + ' '
    
    
    return spelt_gram_string
```

```python
spelt_grams = []
counter = 0
for value in df.cleaned.values:
    spelt_grams.append(two_grammer(value))
    counter += 1
    #if counter %100 == 0:
    print(counter, end = ', ')
```
